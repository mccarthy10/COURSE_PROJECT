# BAR GRAPH DISTRIBUTION OF VOLATILE ACIDITY


## Introduction

This report shows the relationship between the volatile acidity and the quality through a bar chart.The volatile acidity is measured on the vertical-y-axis and the quality on the horizontal-x-axis.The volatile acidity starts reading from 0.0 through to 1.6 while the quality reads from 0 through to 8.

## Body

From the graph,it is shown that the quality of the red wine does not depend on the volatile acidity.Thus the relationship between the quality of the wine and the volatile acidity is not directly proportional.This is because the highest level of volatile acidity measured is correspondent to the lowest level of the quality of the red wine . Therefore,it's observed that there's no pattern or logic with the plotting since not all the values obey the same logic.

## Conclusion

To conclude,the bar chart virtualizes the quality of the red wine and the volatile acidity of the wine.The volatile acidity does not necessarily affect the quality of the red wine.
